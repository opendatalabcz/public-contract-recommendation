```python
def print_gps(gpss):
    for gps in gpss:
        print("{:.5f},{:.5f}".format(gps[0],gps[1]))
```

```python
def generate_random_gps(n, latc=49.743750, lonc=15.338639, lats=1, lons=2):
    res = []
    for _ in range(n):
        lat = numpy.random.normal(latc,lats)
        lon = numpy.random.normal(lonc,lons)
        res.append((lat, lon))
    return res
```

```python
target = numpy.array(generate_random_gps(100))
vectors = numpy.array(generate_random_gps(1000))
```

```python
print_gps(vectors)
```

```python
%%time

res = numpy.zeros((len(target), len(vectors)))
for i, t in enumerate(target):
    for j, v in enumerate(vectors):
        res[i,j]=geo_distance(t, v).km
res
```

```python
%%time

res2 = numpy.zeros((len(target), len(vectors)))
for i, t in enumerate(target):
    for j, v in enumerate(vectors):
        res2[i,j]=great_circle(t, v).km
res2
```

```python
numpy.sum(abs(res - res2))/numpy.sum(res)
```
