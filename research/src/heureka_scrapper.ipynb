```python
category_urls = [
    'https://elektronika.heureka.cz/',
    'https://bile-zbozi.heureka.cz/',
    'https://dum-zahrada.heureka.cz/',
    'https://chovatelstvi.heureka.cz/',
    'https://auto-moto.heureka.cz/',
    'https://detske-zbozi.heureka.cz/',
    'https://moda.heureka.cz/',
    'https://filmy-hudba-knihy.heureka.cz/',
    'https://kosmetika-zdravi.heureka.cz/',
    'https://sport.heureka.cz/',
    'https://hobby.heureka.cz/',
    'https://jidlo-a-napoje.heureka.cz/',
    'https://stavebniny.heureka.cz/',
    'https://sex-erotika.heureka.cz/'
]
```

```python
url = category_urls[0]
category_page = requests.get(url)
category_page
```

```python
category_soup = BeautifulSoup(category_page.text, 'html.parser')
category_soup
```

```python
subcategory_urls = []
for catlist in category_soup.find_all('ul',{'class': 'catlist'}):
    for a in catlist.find_all('a'):
        subcategory_urls.append(a['href'])
subcategory_urls
```

```python
url = subcategory_urls[0]
subcategory_page = requests.get(url)
subcategory_soup = BeautifulSoup(subcategory_page.text, 'html.parser')
subcategory_soup
```

```python
subsubcat_urls = []
for catlist in subcategory_soup.find_all('ul',{'class': 'catlist'}):
    for a in catlist.find_all('a'):
        subsubcat_urls.append(a['href'])
subsubcat_urls
```

```python
url = subsubcat_urls[0]+'?o=1'
subsubcat_page = requests.get(url)
subsubcat_soup = BeautifulSoup(subsubcat_page.text, 'html.parser')
subsubcat_soup
```

```python
products = []
for product in subsubcat_soup.find_all('div',{'class':'product-container'}):
    products.extend(product.find('a').contents)
```
